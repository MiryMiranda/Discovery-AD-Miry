```python
import pandas as pd
df = pd.read_csv('../input/dataset/events_log.csv')
```

#1

a) What is our daily overall clickthrough rate?

```python
# Entries where user visited any of the displayed search results
ct_df = df[df['action'] == 'visitPage']

# overall click-through rate
overall_ct_rate = len(ct_df) / len(df) *100
print('over all click-through rate is : {:.3f}'.format(overall_ct_rate))
```

b) How does it vary between the groups?

```python
for group_name, group_df in df.groupby(by='group'):
    ct_group_df = group_df[group_df['action'] == 'visitPage']
    group_ct_rate = len(ct_group_df) / len(group_df) *100
    print('Group click-through rate for Group ({}) is : {:.3f}'.format(group_name, group_ct_rate))
```

#3

a) What is our daily overall zero results rate?

```python
oz_df = df[df['n_results'] == 0.0]

oz_rate = len(oz_df) / len(df) * 100
print('over all zero result rate is : {:.3f}'.format(oz_rate))
```

How does it vary between the groups?

```python
for group_name, group_df in df.groupby(by='group'):
    oz_group_df = group_df[group_df['n_results'] == 0.0]
    group_oz_rate = len(oz_group_df) / len(group_df) * 100
    print('Group Zero-result rate for Group ({}) is : {:.3f}'.format(group_name, group_oz_rate))
```
